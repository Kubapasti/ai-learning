```python
def divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Error: cannot divide by zero")
        return None
    except TypeError:
        print("Error: both values must be numbers")
        return None
    finally:
        print("divide() finished")  # always runs, like PL/SQL FINALLY

print(divide(10, 2))
print(divide(10, 0))
print(divide(10, "x"))
```

```python
from datetime import datetime

def analyse_account(account):
    try:
        name    = account["name"]
        balance = account["balance"]
        products = account["products"]

        if balance > 50000:
            tier = "Premium"
        elif balance > 10000:
            tier = "Standard"
        else:
            tier = "Basic"

        summary = f"""
        --- Account Summary ---
        Client   : {name}
        Tier     : {tier}
        Balance  : {balance:,.2f} CHF
        Products : {', '.join(products)}
        Generated: {datetime.now().strftime("%Y-%m-%d %H:%M")}
        """
        return summary

    except KeyError as e:
        return f"Error: missing field {e}"

account = {
    "name": "Kuba Pasti",
    "balance": 75000,
    "products": ["current account", "mortgage", "custody"]
}

print(analyse_account(account))
print(analyse_account({"name": "Anna"}))  # missing fields — triggers exception
```

```python

Build an Avaloq client analyser
Write a function called analyse_clients that:

Takes a list of client dictionaries as input
Each client has: name, balance, products (list), country
For each client, assigns a tier — Premium (>50k), Standard (10k–50k), Basic (<10k)
Returns a summary string for each client with all their details
Handles missing fields gracefully with try/except
At the end prints a total count of clients per tier
```

```python


#dupa

def analyse_clients(clients):
    tier_counts = {"Premium": 0, "Standard": 0, "Basic": 0}
    summaries = []

    for client in clients:
        try:
            name = client["name"]
            balance = client["balance"]
            products = client["products"]
            country = client["country"]

            if balance > 50000:
                tier = "Premium"
            elif balance > 10000:
                tier = "Standard"
            else:
                tier = "Basic"

            tier_counts[tier] += 1

            summary = f"""
            --- Client Summary ---
            Name     : {name}
            Tier     : {tier}
            Balance  : {balance:,.2f} CHF
            Products : {', '.join(products)}
            Country  : {country}
            """
            summaries.append(summary)

        except KeyError as e:
            summaries.append(f"Error: missing field {e} for client {client.get('name', 'Unknown')}")

    # Print tier counts
    print("Client Tier Counts:")
    for tier, count in tier_counts.items():
        print(f"{tier}: {count}")

    return summaries


clients = [
    {"name": "Kuba Pasti", "balance": 75000, "products": ["custody", "mortgage"], "country": "PL"},
    {"name": "Anna Kowal", "balance": 12000, "products": ["current account"], "country": "CH"},
    {"name": "John Smith", "balance": 4000, "products": ["savings"], "country": "UK"},
    {"name": "Bad Data", "balance": 99000},  # missing fields — should be handled
]

print("\n".join(analyse_clients(clients)))
```
